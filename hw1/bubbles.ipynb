```python
import pandas as pd

allroutes = pd.read_excel("routes.xlsx",sheet_name = "Sheet1",engine = "openpyxl")
demand_delivery = pd.read_excel("data-demands-deliveyconstraint.xlsx",sheet_name = "Sheet1",engine = "openpyxl")
demand_stop1 = demand_delivery.copy()
demand_stop1.columns = ["Stores_1","Demands_Stop 1","Delivery Contraints_Stop 1"]
demand_stop2 = demand_delivery.copy()
demand_stop2.columns = ["Stores_2","Demands_Stop 2","Delivery Contraints_Stop 2"]
combined_data = pd.merge(allroutes,demand_stop1, how="left", left_on = "stop1", right_on = "Stores_1")
data = pd.merge(combined_data,demand_stop2, how="left", left_on = "stop2", right_on = "Stores_2")
data = data.drop(columns=['Stores_1', 'Stores_2'])
data
```

```python
def route_stop(route,stop):
    if str(stop) == "nan":
        return None
    else:
        return "R"+str(route)+"_"+str(stop)+"_"
data["Route_Stop 1"] = data.apply(lambda x: route_stop(x['Route#'], x['stop1']), axis=1)
data["Route_Stop 2"] = data.apply(lambda x: route_stop(x['Route#'], x['stop2']), axis=1)
routevariables = list(data["Route_Stop 1"])
routevariables = routevariables + list(data["Route_Stop 2"].dropna())
counter_day = 1
variables = []

while counter_day <= 7:
    route_counter = 1
    while route_counter <= len(routevariables):
        x = "D_Day" + str(counter_day)
        x = x +"_"+routevariables[route_counter-1]
        variables.append(x)
        route_counter = route_counter + 1
    counter_day = counter_day + 1
```

```python
stores = list(demand_delivery["Stores"])
days = ["Day1","Day2","Day3","Day4","Day5","Day6","Day7"]
routes = list(allroutes["Route#"])

import gurobipy as grb

#Setting Model
model = grb.Model('Transportation')

#Setting Variables

counter = 1
while counter <= len(variables):
    vars()[variables[counter-1]] = model.addVar(name = variables[counter-1])
    counter = counter + 1

# Setting Objective
obj_var = []
for x in range(len(variables)):
    obj_var.append(vars()[variables[x]])
model.setObjective(sum(obj_var), sense = grb.GRB.MINIMIZE)

# Setting Shop Week Demand Constraint
for x in stores:
    store_var = [k for k in variables if (x+"_") in k]
    store_demand = float(demand_delivery[demand_delivery["Stores"]== x ]["Demands (per week)"])
    con_var = []
    for y in range(len(store_var)):
        con_var.append(vars()[store_var[y]])
    model.addConstr(sum(con_var) >= store_demand, name = ("Demand of "+x))

# Setting Shop Delivery Accepting Constraint
for d in days:
    for s in stores:
        store_var = [k for k in variables if ((s+"_") in k )and(d in k)]
        store_accept = float(demand_delivery[demand_delivery["Stores"]== s ]["Delivery accepting constraints"])
        con_var = []
        for y in range(len(store_var)):
             con_var.append(vars()[store_var[y]])
        model.addConstr(sum(con_var) <= store_accept, name = ("Accept Demand of "+s+" on "+d))
        
# Setting Truck Constraint
for d in days:
    for r in routes:
        route_var = [k for k in variables if (("R"+str(r)+"_") in k )and(d in k)]
        con_var = []
        for y in range(len(route_var)):
             con_var.append(vars()[route_var[y]])
        model.addConstr(sum(con_var) <= 50.0, name = ("Truck capacity of route "+str(r)+" on "+d))

# Non-negative Constraint
for x in variables:
    model.addConstr(vars()[x] >= 0.0, name = ("Non-negative constraint of "+x))

    
#Optimising Step
model.optimize()


print("-------------------------------------")
solution_name = []
solution_value = []

# Print Optimal Solution
for v in model.getVars():
    print('%s = %d'%(v.varName, v.x))
    solution_name.append(v.varName)
    solution_value.append(v.x)

print("-------------------------------------")
    
# Print Optimal Value
print('Objective value: %g' % model.objVal)
```

```python
solution_df = pd.DataFrame(list(zip(solution_name,solution_value)),columns = ["Variable","Value"])
solution_df = solution_df[solution_df["Value"] != 0]
solution_df[["Day","Route","Shop"]] = solution_df["Variable"].str.split('_', expand=True)[[1,2,3]]
solution_df = solution_df[["Day","Route","Shop","Value"]]
solution_df
```
